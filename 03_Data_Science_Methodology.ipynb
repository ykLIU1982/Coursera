I would complete the stages for the spam email detection problem following the data science methodology:  
1) Analytic Approach: Since the question is to detect spam emails, it is a Yes/No question regarding an incoming email. I will select the decision tree classification approach. By checking multiple features related with spam emails, the decision is made by the end on a new email if it is a spam or not.  
2) Data Requirements: The target data content, format, and representation are centered around the features of spam emails. Some required data include but not limited to the sender's address, sending time, the message type (broadcast, cc), sender's signature, and any sensitive words that are common in phishing letters.
