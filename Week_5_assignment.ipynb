#### Add four columns to indicate the count of each dining period

Here, we use a rough classification method to divide the 24 hours of a day into four seperate dining periods: breakfast, lunch, supper, and night.
* breakfast (B): 05:00 - 11:00
* lunch (L): 11:00 - 4:00
* supper (S): 4:00 - 22:00
* night (N): 22:00 - 05:00 (day+1)

Since the turntile data is collected every 4 hours, if the collection period happens to cross two dining periods, the collected passenger number will be proportionally allocated to the counts of these two periods.

```python
df_time_div['Ratio']=df_time_div[['Breakfast', 'Lunch', "Supper", 'Night']].apply(lambda x: [x.Breakfast, x.Lunch, x.Supper, x.Night], axis=1)
```

```python
dict_time_ratio = {v: k for v, k in enumerate(df_time_div.Ratio.tolist())}
```

```python
dict_time_ratio[7]
```

```python
def time_to_ratio(tm):
    HH, MM, SS = (int(x) for x in tm.split(':'))
    hh = int(HH)
    mm = int(MM)
    if mm > 30:
        hh += 1
    hh = hh % 24 # e.g., 08:45 -> 09:00
    return dict_time_ratio[hh]
```

```python
df_tt_count["Ratio"]=df_tt_count.TIME.apply(time_to_ratio)
```

```python
df_tt_count.head()
```

```python
df_tt_count["HourCounts"]=df_tt_count[["PASSENGERS", "Ratio"]].apply(lambda x: [round(x.PASSENGERS*i) for i in x.Ratio], axis=1)
```
